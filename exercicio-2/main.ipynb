## Questão 2

```python
import matplotlib.pyplot as plt
import networkx

graph = networkx.erdos_renyi_graph(n=3000, p=0.001)
networkx.write_gml(graph, "erdos_graph.gml")
```

### 2.a

Vai de acordo com a rede criada e observada no Gephi, pois os esperado esta próximo do gerado

```python
nodes = networkx.number_of_nodes(graph)
links = networkx.number_of_edges(graph)

media_links = 0.001 * (nodes * (nodes - 1) / 2)

print(f"Número de links esperados: {media_links}")
print(f"Total linkks: {networkx.number_of_edges(graph)}")
```

### 2.b

```python
print(networkx.is_directed(graph))
grau_medio = links / nodes
print(grau_medio)
```

Está no regime Supercritical, pois o grau medio (k) é maior que 1
